## 2.1 Loading a file:

Reading file line by line:

```
infile = open("dracula.txt", "r")

load_only = 5
for line in infile:
    print(">>",line, "<<") # notice that the "newline" is still in the line
    #print(">>",line.strip(), "<<") # This can strip these away

    #print(line.split()) # We can now split it word by word

    load_only = load_only - 1
    if load_only < 0:
      break

infile.close()
```

Reading file in one big string:

```
infile = open("dracula.txt", "r")
the_whole_file_as_string = infile.read()
infile.close()
```

## 2.2 Saving into a file

Let's imagine we somehow analyzed the contents of the text document and we want to save it:

```
# For example counting words:
print(the_whole_file_as_string.count("Dracula"))
print(the_whole_file_as_string.count("DRACULA"))
print(the_whole_file_as_string.count("dRaCuLa"))
```

```
# Now its better:
as_lower_case = the_whole_file_as_string.lower()

print(as_lower_case.count("dracula"))
print(as_lower_case.count("count"))

message = "There are "+str(as_lower_case.count("dracula"))+" references to Dracula in the text and "+str(as_lower_case.count("count"))+" references of Counts."
print(message)
```

Amazing analytical task done, now we want to send this research output to our mates ... how can we do it? Saving to file:

```
outfile = open("saved_file.txt", "w")

outfile.write(message)

outfile.close()
```

# 3 (*later in the class*) Help with your text analysis

Simple count (in the whole text):

```
print(as_lower_case.count("dracula"))
```

Over paragraphs:

```
paragraphs = as_lower_case.split("\n\n")

occurrences_list = []

for paragraph in paragraphs:
	count = paragraph.count("dracula")
	occurrences_list.append(count)

print(occurrences_list)
```

Wordlists:

```
tweet_text = "@Holly_Looya @divermam Ok. Ben--Congratulations on your great interview with @IngrahamAngle discussing the disgusting story she made up facts and a total joke compared to @HuffingtonPost. http://somelink.com/"

wordlist = ["made up", "wall", "wow", "joke"]
count = 0
for check_word in wordlist:
    if check_word in tweet_text:
        count = count + 1

print("Text contains", count, "flagged words!")
```

*PS: text analysis is not that simple - coming up with the words in word list might be tricky*

Last example, removing generic words:

```
long_text_example = as_lower_case[10499:11000]
print(long_text_example)
```

```
generic_words_wordlist = ["the", "a", "to", "at", "in", "it", "on", "is", "i", "am", "my", "then", "for", "and", "of", "but", "with", "who", "he", "she", "him", "her"]

for word in long_text_example.split():
    if word.strip() not in generic_words_wordlist:
        print(word)
```
